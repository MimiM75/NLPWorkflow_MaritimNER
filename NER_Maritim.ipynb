**These steps**  are required in the NLP process workflow

*   **Tokenization** getting tokens and remove punctuation (Uni-gram)
*   **Stop word removal** commnly used words
*   **Stemming** based form of words to be analyzed
 as a single term
*   **Lemmitization** grouping the various type of words
*   **POS tagging** tags words to be part of speech (verb, noun, ect)
*   **Info retreival** (NER : Named Entity Recognition)

```python
import nltk
nltk.download('all')
#nltk.download('punkt')
#nltk.download('stopwords')
```

```python
with open('/content/brown_corpus_ca10.txt','r') as myfile:
  doc = myfile.read().replace('\n','')
  doc = doc.replace('/','')
```

# Tokenization

```python
#tokenize
tokens = nltk.word_tokenize(doc)
print("Tokens are : ",tokens)
type(tokens)
```

# Stop word removal

```python
from nltk.corpus import stopwords

#Stop words removing
stopword = stopwords.words('english')
print("Stop words are : ",stopword)

#Remove stop word
for w in tokens:
  if w in stopword:
    tokens.remove(w)
print("Tokens are : ",tokens)
type(tokens)

```

# POS Tagging

```python
tagged_sentence = nltk.pos_tag(tokens)
print("Tagged Tokens are : ",tagged_sentence)
```

# Lemmetization

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("eaten",'v')) #by default it is noun tag
```

```python
from nltk.corpus import wordnet

# Function to convert nltk POS tags to wordnet POS tags
def get_wordnet_pos(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # default to noun if not recognized

# Lemmatize each word with its POS tag
lemmatized_words = []
for word, tag in tagged_sentence:
    pos = get_wordnet_pos(tag)
    lemmatized_word = lemmatizer.lemmatize(word, pos=pos)
    lemmatized_words.append(lemmatized_word)


print("\nLemmatized sentence:")
print(lemmatized_words)
type(lemmatized_words)
```

```python
tagged_sentence = nltk.pos_tag(lemmatized_words)
print("Tagged Tokens are : ",tagged_sentence)
```

## Adding Maritim NER

```python
# Corpus
doc = "I am Captain John, commanding the vessel Ocean Voyager. We recently set sail from the port of Marseille, navigating through the vast sea towards the Atlantic Ocean. The ship is equipped with advanced maritime technology to ensure a safe and efficient journey. Our destination is the bustling port of New York, where we will deliver goods from Europe. Working at sea has always been my dream, and I am proud to be part of such a vital industry."

# Define maritime-specific terms
maritime_terms = ["maritime", "ship", "vessel", "sea", "ocean", "port"]

# Tokenize the Corpus
tokens = nltk.word_tokenize(doc)

# Remove stopwords
stop_words = set(stopwords.words('english'))
tokens = [w for w in tokens if w.lower() not in stop_words]

# POS tagging
tagged_sentence = nltk.pos_tag(tokens)

# Chunk the tagged sentence
ne_chunked_sents = nltk.ne_chunk(tagged_sentence)

# Extract named entities and add maritime-specific terms
ne = []
for tagged_tree in ne_chunked_sents:
    if hasattr(tagged_tree, 'label'):
        entity_name = ' '.join(c[0] for c in tagged_tree.leaves())
        entity_type = tagged_tree.label()
        if any(word.lower() in maritime_terms for word in entity_name.split()):
            ne.append((entity_name, entity_type))
    else:
        # Check if the token is a maritime-specific term
        word, pos = tagged_tree
        if word.lower() in maritime_terms:
            ne.append((word, "MARITIME_TERM"))

print(ne)
```
